```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Download the necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Define the text data
data = "My name is Om Gade and studying AI&DS Engineering at DIT."
print("1. Before converting to lower case: \n  ", data)

# convert to lowercase
data = data.lower()
print("\n2. After converting to lower case: \n  ", data)


# Tokenize the text
word_tokens = word_tokenize(data)
print("\n3. Scentnce converted to tokens: \n  ", word_tokens)

# Get stop words
stop_words = set(stopwords.words('english'))
#print(stop_words)


# Stop word removal with printing removed words
filtered_sentence = []
removed_words = []

for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
    else:
        removed_words.append(w)

# Print the results
print(f"\n4. Stop word removal: \n   { filtered_sentence}")
print(f"\n5. Removed words: \n   {removed_words}")


#Case sensitive stop word removal
#filtered_sentence = [w for w in word_tokens if w not in stop_words]
#print(f"Case sensitive stop word removal: \n{filtered_sentence}")

# Initialize the Porter Stemmer
ps = PorterStemmer()

# Sample words for stemming
sample = ['program', 'programmer', 'programming', 'programs', 'programmers']
print("\n6. Stemming Example")
for s in sample:
    print(f"   {s} : {ps.stem(s)}")

# Example sentence for stemming
sentence = "Programmers program with different programming languages"
words = word_tokenize(sentence)
print("\n7. After tokenization:\n", words)

# Stemming each word in the sentence
print("\n8. Stemming of the scentence  ")
for w in words:
    print(f"{w} : {ps.stem(w)}")
```
